```python
df=pd.read_csv('Titanic-Dataset.csv')
```

```python
df
```

```python
sns.histplot(data="Fare",bins=10,color="green",edgecolor="red")
plt.title("Distribution of Passengers using Histogram")
plt.show()
```

```python
sns.boxplot(x='Sex', y='Age', data=df, color="blue",hue='Survived')
plt.title("Distribution of Passengers using Boxplot")
plt.xlabel=('Sex')
plt.ylabel=('Age')
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Fare')
plt.title("Distribution of Passengers using Scatterplot")
plt.xlabel=('Age')
plt.ylabel=('Fare')
plt.show()
```

```python
sns.distplot(df['Fare'])
```

```python
sns.boxplot(df['Fare'])
```

```python
sns.jointplot(x='Age',y='Fare',data=df)
```

```python
sns.pairplot(df)
```

```python
sns.barplot(data=df,x='Sex',y='Age',color='pink',edgecolor='blue')
plt.title("Distribution of Passengers using barplot")
plt.xlable=('Sex')
plt.ylable=('Age')
plt.show()
```

```python
sns.countplot(x=df['Sex'])
```

```python
sns.violinplot(data=df["Age"],color='purple')
```

```python
sns.violinplot(data=df["Sex"],color='skyblue',edgecolor='red')
```

```python
sns.violinplot(data=df,x="Sex",y="Age",color='purple')
plt.xlable=('Sex')
plt.ylable=('Age')
plt.show()
```

```python
sns.stripplot(data=df,x="Sex",y="Age",color='green')
```

```python
sns.swarmplot(data=df,x="Sex",y="Age",color="blue",edgecolor='red')
plt.title("Distribution of Passengers using Swarmplot")
plt.xlable=('Sex')
plt.ylable=('Age')
plt.show()
```
